```python
df['Categories'] = df['listed_in'].str.split(',')
```

```python
df.head()
```

```python
#Create a list of each occurrence of the categories
#Note two for loops were used as categories result is a list
#.strip() is used to remove white spaces
categories_list = []
for category in df['Categories']:
    for x in category:
        categories_list.append(x.strip())
print(categories_list)
    
```

```python
#Create a unique list of categories
unique_categories = set(categories_list)
unique_categories
```

```python
len(unique_categories)
```

## There are 42 different categories 

```python
#Count number of highest category
counter = 0
categories_dict = {}

for x in categories_list:
    if x == categories_dict.keys():
        counter = counter+1
    else:
        counter = 0


for category in unique_categories:
    categories_dict[category] = counter

```

```python
categories_dict
```

```python
#Create a dictionary with the key as the unique category
#and the value as the number of times is appears
#zip function combines two lists into tuple
#list.count(x) method counts the number of times x appears in the list
#creating a dcitionary with dict(key,value)

my_dict = dict(zip(categories_list, [categories_list.count(i) for i in categories_list]))
```

```python
list_of_actors = df['cast'].str.split(',')
```

```python
list_of_actors
```

```python
#Remove rows with no names
list_of_actors.dropna(how='all',inplace=True)
```

```python
#Create list of actors
actors_list = []
for x in list_of_actors:
    for name in x:
        actors_list.append(name.strip())
```

```python
actors_list
```

```python
actors = dict(zip(actors_list,[actors_list.count(i) for i in actors_list]))
```

```python
actors
```

```python
actors_df = pd.DataFrame(actors,index=['Number'])
```

```python
#Transpose thedataframe and sort from highest value to lowest
actors_df.T.sort_values(by='Number',ascending=False)
```
